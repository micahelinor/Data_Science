## Data

```r
data = data.frame(x1=c(2.5, 2.0, 2.0, 2.1, 8.0, 2.0, 10.0, 10.5, 3.0, 2.5, 3.0, 8.0, 1.0, 1.5),
                  x2=c(5.0, 5.0, 5.0, 4.0, 4.0, 5.5,  6.0,  3.0, 6.0, 6.0, 6.1, 6.0, 1.0, 6.0))
fmt(data)
```

## Exercise 1

Estimate density of x1 distribution.

Plot the estimated density of the x1 distribution, using a Gaussian kernel at bandwidth 1, scaled horizontally 0 to 10.  

You may want to use these function(s):

* geom_density()

```r
out(8,2.4)
ggplot(data) + xlim(0,10) +
geom_density(aes(x1), kernel="gaussian", bw=1)
```

## Exercise 2 

Compare density estimates of the x1 variable distribution at various bandwidths.  

Plot the estimated densities of the x1 variable distribution, using a Gaussian kernel at bandwidths 2, 1, 0.5, and 0.2, overlayed and color-coded by bandwidth, scaled horizontally 0 to 10.

You may want to use these function(s):
* geom_density()

Use `geom_density(..., alpha=0.3))` to make the fill transparent.

## Exercise 3

Plot the estimated density of x1 and x2 distributions taken together, using a Gaussian kernel at bandwidths 10, 5, and 1, scaled horizontally and vertically 0 to 10, one plot for each bandwidth.  Show the original observations upon which the density is estimated.

You may want to use these function(s):
* geom_density_2d_filled()
* geom_density_2d()

```r
out(2.5,2.5)
ggplot(data) + xlim(0,10) + ylim(0,10) +
geom_density_2d_filled(aes(x=x1, y=x2), kernel="gaussian", h=10, bins=100, show.legend=FALSE) +
geom_density_2d(aes(x=x1, y=x2), kernel="gaussian", h=10, bins=20, size=0.2) +
geom_point(aes(x=x1, y=x2), color="black") +
scale_fill_manual(values=heat.colors(n=100, alpha=0.8))
```

## Exercise 4

Determine the probability that a new observation will measure 1 < x1 < 3. Base the probability on a kernel density estimate using Gaussian kernel, standard deviation 1, estimated over the range 0 to 10.

You may want to use these function(s):
* density()
* approxfun()
* integrate()

```r
pdf = approxfun(density(data$x1, kernel="gaussian", bw=1, from=0, to=10))

integral.1.3 = integrate(pdf, 1, 3)$value
layout(fmt(integral.1.3, "prob.1.3"))
```
