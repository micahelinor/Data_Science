## Data

```r
data = data.frame(x1=c(2.5, 2, 2, 6.5, 8, 7.5, 10, 10.5),
                  x2=c(9, 3.5, 1, 11, 2.5, 8.5, 4, 7.5),
                  x3=c(9, 10, 1, 2, 1.5, 10.5, 1, 11),
                  x4=c(9, 9, 9, 10, 10, 10.5, 9.5, 11),
                  x5=c(3, 7, 6.5, 1, 10, 10, 1, 1))
fmt(data)
```

## Problem 1

For each variable in the dataset, show the standard deviation, variance, cumulative variance (progressing through x1, x2, x3, x4, x5), relative variance, and cumulative relative variance (progressing through x1, x2, x3, x4, x5).  Show a scree plot of the dataset. Show a 3D scatterplot of the dataset for the 3 variables with the most variation.

You might want to use these function(s):
* sd()
* variance()
* cumsum()
* sum()
* ylim()
* geom_col()
* plot3d()

Use `plot3d(data[,c("x3","x5","x2")], ..., radius=0.4, ...)` to specify the points in the 3D scatterplot.

```r
variable = c("x1", "x2", "x3", "x4", "x5")
standard_deviation = c(sd(data$x1), sd(data$x2), sd(data$x3), sd(data$x4), sd(data$x5))
variance = c(var(data$x1), var(data$x2), var(data$x3), var(data$x4), var(data$x5))
cum_variance = cumsum(variance)
relative_variance = variance / sum(variance)
cum_relative_variance = cumsum(relative_variance)

scree = data.frame(variable, standard_deviation, variance, cum_variance, relative_variance, cum_relative_variance)
fmt(scree, "scree of data")
```

```r
out(5,2.5)
ggplot(scree) + ylim(0,1) + geom_col(aes(x=variable, y=relative_variance)) + theme.no_legend
```

## Problem 3

Transform the **normalized** dataset to principal component representation.  Show the centroid and weight matrix used to calculate the principal components.  Show the transformed dataset.  For each variable in the transformed dataset, show the standard deviation, variance, cumulative variance (progressing through PC1, PC2, PC3, PC4, PC5), relative variance, and cumulative relative variance (progressing through PC1, PC2, PC3, PC4, PC5).  Show a scree plot of the transformed dataset. Show a 3D scatterplot of the transformed dataset for the 3 variables with the most variation.

You might want to use these function(s):
* prcomp()

Use `prcomp(data.norm, scale=FALSE, retx=TRUE)` to transform the dataset.  Set `scale=FALSE` because the dataset is already normalized - it does not have to be normalized again.  Use `plot3d(data.norm[,c("PC1","PC2","PC3")], ..., radius=0.15, ...)` to specify the points in the 3D scatterplot.

```r
pc = prcomp(data.norm, scale=FALSE, retx=TRUE)
fmt(pc$center, "pc$center", row.names=TRUE)
```

```r
fmt(pc$rotation, "pc$rotation", row.names=TRUE)
```

```r
data.norm.pc = as.data.frame(pc$x)
fmt(data.norm.pc, "data.norm.pc", row.names=FALSE)
```

```r
variable = c("PC1", "PC2", "PC3", "PC4", "PC5")
standard_deviation = c(sd(data.norm.pc$PC1), sd(data.norm.pc$PC2), sd(data.norm.pc$PC3), sd(data.norm.pc$PC4), sd(data.norm.pc$PC5))
variance = c(var(data.norm.pc$PC1), var(data.norm.pc$PC2), var(data.norm.pc$PC3), var(data.norm.pc$PC4), var(data.norm.pc$PC5))
cum_variance = cumsum(variance)
relative_variance = variance / sum(variance)
cum_relative_variance = cumsum(relative_variance)

scree = data.frame(variable, standard_deviation, variance, cum_variance, relative_variance, cum_relative_variance)
fmt(scree, "scree of data.norm.pc")
```

```r
out(5,2.5)
ggplot(scree) + ylim(0,1) + geom_col(aes(x=variable, y=relative_variance)) + theme.no_legend
```

## Problem 4

Transform the new observation [x1=5, x2=4, x3=3, x4=10, x5=1] to principal component representation, where the principal components are calculated based on the transformed dataset without the new observation.  Show a new dataset comprising the transformed dataset and the transformed new observation.  Show a 3D scatterplot of this new dataset for the first 3 principal components, and distinguish the new observation from the other observations by color.

You might want to use these function(s):
* predict()
* rbind()
* plot3d()

Use `plot3d(data.combined[,c("PC1","PC2","PC3")], ..., radius=0.15, col=PALETTE[c(1,1,1,1,1,1,1,1,2])` to specify colors of the points in the 3D scatterplot.

```r
new = data.frame(x1=5, x2=4, x3=3, x4=10, x5=1)
new.norm = data.frame(x1=(new$x1-mean(data$x1)) / sd(data$x1),
                      x2=(new$x2-mean(data$x2)) / sd(data$x2),
                      x3=(new$x3-mean(data$x3)) / sd(data$x3),
                      x4=(new$x4-mean(data$x4)) / sd(data$x4),
                      x5=(new$x5-mean(data$x5)) / sd(data$x5))

new.norm.pc = data.frame(PC1=new.norm$x1*pc$rotation["x1","PC1"] + new.norm$x2*pc$rotation["x2","PC1"] + new.norm$x3*pc$rotation["x3","PC1"] + new.norm$x4*pc$rotation["x4","PC1"] + new.norm$x5*pc$rotation["x5","PC1"],
                         PC2=new.norm$x1*pc$rotation["x1","PC2"] + new.norm$x2*pc$rotation["x2","PC2"] + new.norm$x3*pc$rotation["x3","PC2"] + new.norm$x4*pc$rotation["x4","PC2"] + new.norm$x5*pc$rotation["x5","PC2"],
                         PC3=new.norm$x1*pc$rotation["x1","PC3"] + new.norm$x2*pc$rotation["x2","PC3"] + new.norm$x3*pc$rotation["x3","PC3"] + new.norm$x4*pc$rotation["x4","PC3"] + new.norm$x5*pc$rotation["x5","PC3"],
                         PC4=new.norm$x1*pc$rotation["x1","PC4"] + new.norm$x2*pc$rotation["x2","PC4"] + new.norm$x3*pc$rotation["x3","PC4"] + new.norm$x4*pc$rotation["x4","PC4"] + new.norm$x5*pc$rotation["x5","PC4"],
                         PC5=new.norm$x1*pc$rotation["x1","PC5"] + new.norm$x2*pc$rotation["x2","PC5"] + new.norm$x3*pc$rotation["x3","PC5"] + new.norm$x4*pc$rotation["x4","PC5"] + new.norm$x5*pc$rotation["x5","PC5"])

fmt(new.norm.pc, "new.norm.pc", row.names=FALSE)
```

```r
data.combined = rbind(data.norm.pc, new.norm.pc)
fmt(data.combined, "data.combined", row.names=FALSE)
```
